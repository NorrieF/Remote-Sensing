```python
import plotly.express as px
import pandas as pd

# load the data
df = pd.read_csv(r'C:\Users\Norrie\Desktop\ee-chart.csv', parse_dates=['Date'])

# plot the data
fig = px.scatter(df, x = 'Date', y = 'NDVI', color = 'NDVI', hover_name = 'Date', color_continuous_scale = px.colors.diverging.RdYlGn, template='plotly_dark')
fig.show()

# export the figure in a html file
import plotly.io as pio
pio.write_html(fig, file = "index.html", auto_open = True)
```
